```
import csv
import os
import tensorflow as tf
import numpy as np
from keras import Model
from keras.layers import Input, Dense
```

```
class LinearRegression(Model):
  def __init__(self):
    super(LinearRegression,self).__init__()
    self.dense = Dense(1,input_dim=10)
    
  def call(self,x):
    return self.dense(x)
```

```
df_new = df.dropna(subset=['Crossing','Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling','Curve', 'FKAccuracy', 'LongPassing', 'BallControl'])
df_new.shape[0]
```

```
x_train = np.array(df_new[['Crossing','Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling','Curve', 'FKAccuracy', 'LongPassing', 'BallControl']])
y_train = np.array(df_new['Overall'])
```

```
reglin = LinearRegression()
reglin.compile(loss='mse')
reglin.fit(x_train ,y_train)
```

```
import random
novas_notas = []

#Generate 5 random numbers between 10 and 30
for i in range(3):
  randomlist = random.sample(range(0, 100),10)
  novas_notas.append(randomlist)

novas_notas
```

```
out = reglin.predict(novas_notas) 
for i in range(0,len(out)):
  print(str(novas_notas[i])+' -> '+str(out[i]))
```
